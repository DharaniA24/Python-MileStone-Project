```python
contact = {}

def display_contact():
    print("Name t\t Contact Number")
    for key in contact:
        # print(key)
        print("{}\t\t{}".format(key, contact.get(key)))

while True:
    try:
        choice = int(input("Please select an option from 1 to 6 \n 1. Add New Contact \n 2. Search Contact \n 3. Display Contact \n 4. Edit Contact \n 5. Delete Contact \n 6. Exit Contact \n Enter your choice: "))
    except ValueError:
        print("Invalid input. Please enter a number from 1 to 6.")
        continue
    
    if choice == 1:
        try:
            name = input("Enter the contact name: ")
            phone = int(input("Enter the mobile number: "))
            contact[name] = phone
            print('\nContact added successfully.\n')
        except Exception as e:
            print(f"An error occurred: {e}\n")
        
    elif choice == 2:
        search_name = input("Search the contact name: ")
        if search_name in contact:
            print(search_name, "'s contact number is ", contact[search_name], '\n')
        else:
            print("Name is not found in contact book \n")
            
    elif choice == 3:
        if not contact:
            print("Empty contact book \n")
        else:
            display_contact()
            print('\n')
            
    elif choice == 4:
        try:
            edit_contact = input("Enter the contact name to be edited: ")
            if edit_contact in contact:
                phone = int(input("Enter new mobile number: "))
                contact[edit_contact] = phone
                print("Contact updated successfully.\n")
                display_contact()
                print('\n')
            else:
                print("Name is not found in contact book \n")
        except Exception as e:
            print(f"An error occurred: {e}\n")
            
    elif choice == 5:
        del_contact = input("Enter the contact name to be deleted: ")
        if del_contact in contact:
            confirm = input("Do you want to delete this contact yes/no (type y or Y)? ")
            if confirm.lower() == "y":
                contact.pop(del_contact)
                print("Contact deleted successfully.\n")
            else:
                print("Deletion cancelled.\n")
        else:
            print("Name is not found in contact book \n")
        
    elif choice == 6:
        print("Exiting contact book!")
        break
        
    else:
        print("Invalid choice. Please select a number from 1 to 6.\n")
```
